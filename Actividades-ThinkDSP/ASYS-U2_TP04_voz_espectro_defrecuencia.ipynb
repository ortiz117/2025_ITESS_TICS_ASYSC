```python
import thinkdsp
import matplotlib.pyplot as plt

# Configurar el tamaño de las gráficas para que se vean mejor
plt.rcParams['figure.figsize'] = (10, 4)

#Cargamos los dos archivos .wav desde la carpeta.
# thinkdsp.read_wave() lee el archivo y crea un objeto 'Wave'.
wave1 = thinkdsp.read_wave('ZAPATO.wav')
wave2 = thinkdsp.read_wave('ZAPATOJ.wav')

#Graficamos la primera onda.
# .plot() en un objeto 'Wave' muestra la amplitud (eje Y) contra el tiempo (eje X).
print("--- Gráficas en el Dominio del Tiempo ---")
plt.subplot(1, 2, 1) # 1 fila, 2 columnas, gráfico 1
wave1.plot()
plt.title('Onda - Palabra 1')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

#Graficamos la segunda onda en el segundo subplot.
plt.subplot(1, 2, 2) # 1 fila, 2 columnas, gráfico 2
wave2.plot()
plt.title('Onda - Palabra 2')
plt.xlabel('Tiempo (s)')

plt.tight_layout() # Ajusta las gráficas para que no se amontonen
plt.show()

#Estas gráficas muestran la forma del sonido de dos formas
#muy parecidas, ya que se dijo la misma palabra. Las pequeñas diferencias
#en amplitud y duración son lo que las hace dos grabaciones únicas.

#Calculamos el espectro de cada onda.
#.make_spectrum() aplica la Transformada de Fourier
#para convertir la señal del dominio del tiempo al dominio de la frecuencia.
spectrum1 = wave1.make_spectrum()
spectrum2 = wave2.make_spectrum()

#Graficamos el primer espectro.
#.plot() en un objeto 'Spectrum' muestra la amplitud (eje Y) contra la frecuencia (eje X).
print("\n--- Gráficas en el Dominio de la Frecuencia ---")
plt.subplot(1, 2, 1) # 1 fila, 2 columnas, gráfico 1
spectrum1.plot()
plt.title('Espectro - Palabra 1')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
#Limitamos el eje X a 5000 Hz, ya que la mayoría de la energía de la voz está ahí.
plt.xlim(0, 5000) 

#Graficamos el segundo espectro.
plt.subplot(1, 2, 2) # 1 fila, 2 columnas, gráfico 2
spectrum2.plot()
plt.title('Espectro - Palabra 2')
plt.xlabel('Frecuencia (Hz)')
plt.xlim(0, 5000) # Usamos el mismo límite para poder comparar

plt.tight_layout()
plt.show()
```
